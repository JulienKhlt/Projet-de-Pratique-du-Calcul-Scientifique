```julia
using Plots
```

Modèle SIR \
Paramètres de l'épidémie :

```julia
β = 1
γ = 0.25
pas_tps = 0.1
tps = 40
```

Conditions initiales :

```julia
S = 99
I = 1
R = 0
```

Fonction équation différentielle ($U' (t) = f(t, U(t))$) :

```julia
function f(U, β, γ, N)
    dUdt = [- β*U[2]*U[1]/N, β*U[2]*U[1]/N - γ*U[2], γ*U[2]]
    return dUdt
end
```

Euler explicite et affichage :

```julia
function epidemie(S, I, R, β, γ, pas_tps, tps)

    Susceptible = Float64[S]
    Infectious = Float64[I]
    Recovered = Float64[R]

    N = S + I + R
    U = Float64[S, I, R]

    for t in range(pas_tps, tps, length = Int(tps/pas_tps))
        dUdt = f(U, β, γ, N)

        U += dUdt * pas_tps

        append!(Susceptible, U[1])
        append!(Infectious, U[2])
        append!(Recovered, U[3])
    end

    T = range(0, tps, length = Int(tps/pas_tps)+1)

    p = plot(T, Susceptible, color="#4a4c4d", linewidth=2.0, linestyle=:solid, label="Susceptible", xlabel="Temps", ylabel="Nombre de personne")
    plot!(p, T, Infectious, color="#c13607", linewidth=2.0, linestyle=:solid, label="Infectious")
    plot!(p, T, Recovered, color="#3c9f66", linewidth=2.0, linestyle=:solid, label="Recovered")
end
```

On lance le programme :

```julia
epidemie(S, I, R, β, γ, pas_tps, tps)
```

Si on fait varier les paramètres :

```julia
γ_range = range(0, 1.5, length=100)

anim = @animate for γ in γ_range
    epidemie(S, I, R, β, γ, pas_tps, tps)
end
gif(anim, "anim.gif", fps = 20)
```

```julia
ß_range = range(0, 2, length=100)

anim2 = @animate for ß in ß_range
    epidemie(S, I, R, ß, γ, pas_tps, tps)
end
gif(anim2, "anim2.gif", fps = 15)
```

**Montrons qu'il existe une et une seule solution au modèle SIR:**


$E = \mathbb{R}^3$ est un espace de Banach <br>
Notons $\Omega$, un ouvert de $\mathbb{R} \times E$. <br>
La fonction $f$ définie par:

$$
f : \left\{
    \begin{array}{ll}
        \Omega \rightarrow E \\
        (t, U(t)=(S(t), I(t), R(t))) \rightarrow (- \frac{\beta SI}{N}, \frac{\beta SI}{N} - \gamma I, \gamma I)
    \end{array}
\right.
$$
est continue car polynomiale. <br>

<br>
De plus on a :
$$
\begin{array}{ll}
    \left\|f(t, U_1) - f(t, U_2)\right\|_\infty &= \left\|(- \frac{\beta S_1I_1}{N}, \frac{\beta S_1I_1}{N} - \gamma I_1, \gamma I_1) - (- \frac{\beta S_2I_2}{N}, \frac{\beta S_2I_2}{N} - \gamma I_2, \gamma I_2)\right\|_\infty \\
&= \left\|(- \frac{\beta S_1I_1-S_2I_2}{N}, \frac{\beta S_1I_1-S_2I_2}{N} - \gamma (I_1-I_2), \gamma (I_1-I_2)\right\|_\infty \\
\end{array}
$$

Or puisque : $|\frac{\beta S_1I_1-S_2I_2}{N} - \gamma (I_1-I_2)|< max(\gamma |(I_1-I_2)|, |\frac{\beta S_1I_1-S_2I_2}{N}|) $ <br>

Et : $\gamma |(I_1-I_2)|<\gamma \left\|U_1 - U_2\right\|_\infty$ <br>
Sans perte de généralité on peut supposer : $|S_1I_1 - S_2I_2| = S_1I_1 - S_2I_2$ <br>
Alors soit $S_1 \leq S_2$ alors $|S_1I_1 - S_2I_2| = S_1(I_1 - \frac{S_2}{S_1}I_2) \leq S_1(I_1 - I_2)$ <br>
soit $I_1 \leq I_2$ alors $|S_1I_1 - S_2I_2| = I_1(S_1 - \frac{I_2}{I_1}S_2) \leq S_1(I_1 - I_2)$

```julia
Nombre_maille = 100
x = range(0, 1, length=Nombre_maille)
dx = 1/Nombre_maille
D = 0.001
```

```julia
U = [zeros(Nombre_maille) for _ in 1:3]
laplacien = [zeros(Nombre_maille) for _ in 1:3]
```

```julia
for i = 1:Nombre_maille
    U[1][i] = 1
    U[2][i] = 0
    U[3][i] = 0
end

U[2][50] = 1
```

```julia
function f(U, β, γ, N)
    dUdt = [- β*U[2].*U[1]/N, β*U[2].*U[1]/N - γ*U[2], γ*U[2]]
    return dUdt
end
```

```julia
# stable si Ddt/dx^2 =< 1/2

pas_tps = 3e-2
tps = 10
β = 7
γ = 0.25
```

```julia
function diffusion(U, β, γ, D, x, dx, pas_tps, tps)
    Epidemie = [U]
    N = Nombre_maille + 1
    for t in range(pas_tps, tps, length = trunc(Int, (tps/pas_tps)))
        for k in 2:Nombre_maille-1
            for i = 1:length(U)
            laplacien[i][k] = (U[i][k+1]-2*U[i][k]+U[i][k-1])/(dx^2)
            end
        end
        U .+= pas_tps*(D*laplacien + f(U, β, γ, N)) 
        append!(Epidemie, [U])
    end
    for t in 1:trunc(Int, (tps/pas_tps))
        if t%10 == 0
            plot(x, Epidemie[t][1], color="#4a4c4d", linewidth=2.0, linestyle=:solid, label="Susceptible", xlabel="Position", ylabel="Nombre de personne")
            plot(x, Epidemie[t][2], color="#c13607", linewidth=2.0, linestyle=:solid, label="Infectious")
            plot(x, Epidemie[t][3], color="#3c9f66", linewidth=2.0, linestyle=:solid, label="Recovered")
            # title(string("Model Diffusion à t=", string(t)))
        end
    end
    show()
end
```

```julia
diffusion(U, β, γ, D, x, dx, pas_tps, tps)
```
